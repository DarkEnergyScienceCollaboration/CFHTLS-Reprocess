## Same notebook as CheckPhotometry but with photometric calibration with respect to CFHTLS T0007 catalog

```python
%matplotlib inline
import matplotlib
import matplotlib.pylab as plt

from lsst.meas.astrom import LoadAstrometryNetObjectsTask as Astrometry
import lsst.daf.persistence as dafPersist
import lsst.afw.geom as afwGeom
import lsst.afw.cameraGeom as cg
import lsst.afw.table as afwTable
import lsst.afw.image as afwImage
import lsst.afw.coord as afwCoord
import numpy
import math

file_cat = "../../Catalogs/CFHTLS_D3_stars_mag_r_20.cat"
with open(file_cat, 'r') as f:
    lines = f.readlines()
    
schema  = afwTable.SourceTable.makeMinimalSchema()
filters='ugriz'

for f in filters :
    schema.addField(f + "_mag", type=numpy.float32, doc="magnitude" + f)
    schema.addField(f + "_magSigma", type=numpy.float32, doc="magnitude error" + f)
schema.addField("s_flag", type=numpy.int, doc="SExtractor Flag")
schema.addField("t_flag", type=numpy.int, doc="Type Flag")

catalog = afwTable.SourceCatalog(schema)
for i in range(2,len(lines)) :
    l = lines[i].split()
    record = catalog.addNew()
    coord = afwCoord.Coord(afwGeom.Point2D(numpy.float64(l[9]),numpy.float64(l[10])))
    record.set("coord",coord)                  
    for cnts,f in enumerate(filters) :
        record.set(f + "_mag", float(l[11+2*cnts]))
        record.set(f + "_magSigma", float(l[12+2*cnts]))
    record.set("s_flag", int(l[21]))
    record.set("t_flag", int(l[22]))

cfhtRef = catalog.copy(deep=True)
print("Catalog cfhtRef has been loaded with %d CFHT reference objects"%(len(lines)-2))
```

```python
# Validation plots for aperture and psf photometry
# Apterture radii in pixels are : 3.0, 4.5, 6.0, 9.0, 12.0, 17.0, 25.0, 35.0, 50.0, 70.0

%matplotlib inline
import matplotlib
import matplotlib.pylab as plt

from lsst.meas.astrom import LoadAstrometryNetObjectsTask as Astrometry
import lsst.daf.persistence as dafPersist
import lsst.afw.geom as afwGeom
import lsst.afw.cameraGeom as cg
import lsst.afw.table as afwTable
import lsst.afw.image as afwImage
import lsst.afw.display.ds9 as ds9
import numpy
import math


visit = 906675
filter = 'r'

# Do not consider reference magnitude > 19 
magMax = 19

#setup astrometry
solve = Astrometry.ConfigClass()
astrom = Astrometry(solve)

# List of flags for quality checks
Flags = ["base_PixelFlags_flag_saturated", "base_PixelFlags_flag_cr", "base_PixelFlags_flag_interpolated",
         "base_PsfFlux_flag_edge","slot_Centroid_flag", "base_ClassificationExtendedness_flag"]

#setup butler - It assumes that a standard LSST directory structure is setup at the upper directory level
butler = dafPersist.Butler('../output')

diffpsf = []
diffAp = []
mags = []
ccdNum = []
xCoord = []
yCoord = []
magsPsf = []
errsPsf = [] # Error on PSF mag
magsAp = []
errsAp = []

for ccd in range(0,36) :
    vId = {'visit':visit,'ccd':ccd}
    src = butler.get('src', vId, immediate=True)
    md = butler.get("calexp_md", vId, immediate=True)
    calib = afwImage.Calib(md)
    #  match CFHT sources to external reference catalog
    match = afwTable.matchRaDec(cfhtRef, src, 1.0 * afwGeom.arcseconds)
    print("CCD %d : %d matches found"%(ccd,len(match)))
    count = 0
    # Loop over matched sources
    for cnts,m in enumerate(match):
        # Apply quality cuts
        rej = 0
        for f in Flags :
            if m.second.get(f) :
                rej = 1
                break
        if rej == 1 :
            continue
#        if m.first.get("s_flag") != 0 :
#            print "Not photometric"
#            continue
        signalToNoise = m.second.get("base_PsfFlux_flux")/m.second.get("base_PsfFlux_fluxSigma")
        if signalToNoise < 5.0 :
            continue
        # Select star-like sources
        if m.second.get('base_ClassificationExtendedness_value') > 0.5:
            continue
        mag = m.first.get(filter + "_mag")
        
        # only keep magnitude < magMax
        if mag > magMax :
            continue
        
        magPsf, errPsf = calib.getMagnitude(m.second.get('base_PsfFlux_flux'),m.second.get('base_PsfFlux_fluxSigma'))
        diffpsf.append(mag - magPsf)
        
        mags.append(mag)
        magsPsf.append(magPsf)
        errsPsf.append(errPsf)
#        magsPsf.append(m.second.get('base_PsfFlux_flux'))
#        errsPsf.append(m.second.get('base_PsfFlux_fluxSigma'))
        x = m.second.getCentroid().getX()
        y = m.second.getCentroid().getY()
        xCoord.append(x)
        yCoord.append(y)
        ccdNum.append(ccd)
        
        # Iterate over aperture radii
        for ir in range(10) :
            apBase = "base_CircularApertureFlux_" + str(ir)
            apFlux = apBase + "_flux"
            apFluxSig = apBase + "_fluxSigma"
            apFluxFlag = apBase + "_flag"
            diffAp.append([])
            magsAp.append([])
            errsAp.append([])
            if m.second.get(apFluxFlag) :
                # if aperture flux flag is set, set entry to large negative value
                diffAp[ir].append(-100)
                magsAp[ir].append(-100)
                errsAp[ir].append(-100)
                
            else :
                diffAp[ir].append(mag - calib.getMagnitude(m.second.get(apFlux)))
                magAp, errAp = calib.getMagnitude(m.second.get(apFlux),m.second.get(apFluxSig))
                magsAp[ir].append(magAp)
                errsAp[ir].append(errAp)

        count += 1
#    print "ccd : ",ccd," number of matches :",count
        
print "Total number valid of matches :", len(mags)
```

```python
metadata = src.getMetadata()
radii = metadata.get("base_CircularApertureFlux_radii")

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(18, 10))

sigmaPsf = numpy.std(diffpsf)
meanPsf = numpy.mean(diffpsf)
ax1.hist(diffpsf, bins=100, color="blue", range=[-0.4,0.3])
ax1.set_xlabel("$\Delta$ mag (PSFex)", fontsize=20)
ax1.set_title("$\sigma$ = %.0f mmag \nmean = %.1f mmag"%(sigmaPsf*1000, meanPsf*1000), x=0.2, y=0.9, fontsize=20)
ax1.tick_params(labelsize=20)

ax2.scatter(ccdNum, diffpsf, s=8, color="blue")
ax2.set_ylim([-0.15,0.15])
ax2.tick_params(labelsize=20)
ax2.set_xlabel("CCD #", fontsize=20)
ax2.set_ylabel("$\Delta$ mag (PSFex)", fontsize=20)

fig.tight_layout()

fig, (ax3, ax4) = plt.subplots(ncols=2, figsize=(18, 10))
ax3.scatter(diffAp[3],diffpsf, s=5, color="blue")
ax3.set_xlim([-0.15,0.15])
ax3.set_ylim([-0.15,0.15])
ax3.set_xlabel("$\Delta$ mag (Ap = %.1f pix)"%(radii[3]),fontsize=20)
ax3.set_ylabel("$\Delta$ mag (PSFex)",fontsize=20)
ax3.tick_params(labelsize=20)

ax4.scatter(mags,diffpsf, s=5, color="blue")
#ax4.set_xlim([-0.15,0.15])
ax4.set_ylim([-0.15, 0.15])
ax4.set_xlabel("Magnitude (ref)",fontsize=20)
ax4.set_ylabel("$\Delta$ mag (PSFex)",fontsize=20)
ax4.tick_params(labelsize=20)

fig.tight_layout()
```

### Compare aperture flux estimated for different radii to flux in SDSS reference catalog

```python
metadata = src.getMetadata()
radii = metadata.get("base_CircularApertureFlux_radii")

fig, (ax1, ax2) = plt.subplots(nrows= 2, ncols=5, figsize=(18, 15))

for ir in range(10) :
    idxs = numpy.where(numpy.asarray(diffAp[ir]) > -50)
    tmp = numpy.asarray(diffAp[ir])
    mean = numpy.mean(tmp[idxs])
    sigma = numpy.std(tmp[idxs])
    hr = [mean-0.4, mean+0.3]
    if ir < 5 :
        ax1[ir].hist(tmp[idxs], bins=100, range=hr)
        ax1[ir].set_xlabel("$\Delta$ mag - Aperture radius :%.1f pix"%radii[ir])
        ax1[ir].set_title("mag < %.1f \n $\sigma$ = %.0f mmag"%(magMax,sigma*1000), x=0.25, y=0.9)
    else :
        ax2[ir-5].hist(tmp[idxs], bins=100, range=hr)
        ax2[ir-5].set_xlabel("$\Delta$ mag - Aperture radius :%.1f pix"%radii[ir])
        ax2[ir-5].set_title("mag < %.1f \n$\sigma$ = %.0f mmag"%(magMax,sigma*1000), x=0.25, y=0.9)
    
```
