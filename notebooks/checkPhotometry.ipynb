# Check PSF and Aperture photometry on a single visit (1 filter - 36 CCD)

```python
# Validation plots for aperture and psf photometry
# Apterture radii in pixels are : 3.0, 4.5, 6.0, 9.0, 12.0, 17.0, 25.0, 35.0, 50.0, 70.0

%matplotlib inline
import matplotlib
import matplotlib.pylab as plt

from lsst.meas.astrom import LoadAstrometryNetObjectsTask as Astrometry
import lsst.daf.persistence as dafPersist
import lsst.afw.geom as afwGeom
import lsst.afw.cameraGeom as cg
import lsst.afw.table as afwTable
import lsst.afw.image as afwImage
import lsst.afw.display.ds9 as ds9
import numpy
import math


visit = 906675
filter = 'r'

# Do not consider reference magnitude > 19 
magMax = 19

#setup astrometry
solve = Astrometry.ConfigClass()
astrom = Astrometry(solve)

# List of flags for quality checks
Flags = ["base_PixelFlags_flag_saturated", "base_PixelFlags_flag_cr", "base_PixelFlags_flag_interpolated",
         "base_PsfFlux_flag_edge","slot_Centroid_flag", "base_ClassificationExtendedness_flag"]

#setup butler - It assumes that a standard LSST directory structure is setup at the upper directory level
butler = dafPersist.Butler('../output')

# setup camera
cam = butler.get('camera')
ccd14 = cam['ccd14']
vId = {'visit':visit, 'ccd':14}
calexp = butler.get("calexp", vId, immediate=True)
detCenter = ccd14.getCenter(cg.PIXELS).getPoint()
detCenterCoord = calexp.getWcs().pixelToSky(detCenter)

# Retrieve external reference catalog (SDSS DR9)
# loadSkyCircle is replacing getReferenceSources and is returning a structure (fluxField='r_flux'; refCat)
s = astrom.loadSkyCircle(detCenterCoord, afwGeom.Angle(1., afwGeom.degrees), filter)
refCat = s.refCat

diffpsf = []
diffAp = []
mags = []
ccdNum = []
xCoord = []
yCoord = []
magsPsf = []
errsPsf = [] # Error on PSF mag
magsAp = []
errsAp = []

for ccd in range(0,36) :
    vId = {'visit':visit,'ccd':ccd}
    src = butler.get('src', vId, immediate=True)
    md = butler.get("calexp_md", vId, immediate=True)
    calib = afwImage.Calib(md)
    #  match CFHT sources to external reference catalog
    match = afwTable.matchRaDec(refCat, src, 1.0 * afwGeom.arcseconds)
    count = 0
    # Loop over matched sources
    for cnts,m in enumerate(match):
        # Apply quality cuts
        rej = 0
        for f in Flags :
            if m.second.get(f) :
                rej = 1
                break
        if rej == 1 :
            continue
        if not m.first.get("photometric") :
            print "Not photometric"
            continue
        signalToNoise = m.second.get("base_PsfFlux_flux")/m.second.get("base_PsfFlux_fluxSigma")
        if signalToNoise < 5.0 :
            continue
        # Select star-like sources
        if m.second.get('base_ClassificationExtendedness_value') > 0.5:
            continue
        # Correct refereence object magnitude to take into account the different filter sets in SDSS and CFHT
        # Ref : http://www3.cadc-ccda.hia-iha.nrc-cnrc.gc.ca/community/CFHTLS-SG/docs/extra/filters.html
        # Another color correction is given in : 
        # http://www1.cadc-ccda.hia-iha.nrc-cnrc.gc.ca/community/CFHTLS-SG/docs/cfhtls.pdf
        if filter == 'u' :
            mag = afwImage.abMagFromFlux(m.first.get('u_flux')) - 0.241*(afwImage.abMagFromFlux(m.first.get('u_flux')) - afwImage.abMagFromFlux(m.first.get('g_flux')))
        elif filter == 'i' :
            mag = afwImage.abMagFromFlux(m.first.get('i_flux')) - 0.085*(afwImage.abMagFromFlux(m.first.get('r_flux')) - afwImage.abMagFromFlux(m.first.get('i_flux')))
        elif filter == 'r' :
            mag = afwImage.abMagFromFlux(m.first.get('r_flux')) - 0.024*(afwImage.abMagFromFlux(m.first.get('g_flux')) - afwImage.abMagFromFlux(m.first.get('r_flux')))
        elif filter == 'g' :
            mag = afwImage.abMagFromFlux(m.first.get('g_flux')) - 0.153*(afwImage.abMagFromFlux(m.first.get('g_flux')) - afwImage.abMagFromFlux(m.first.get('r_flux')))

        # only keep magnitude < magMax
        if mag > magMax :
            continue
        
        magPsf, errPsf = calib.getMagnitude(m.second.get('base_PsfFlux_flux'),m.second.get('base_PsfFlux_fluxSigma'))
        diffpsf.append(mag - magPsf)
        
        mags.append(mag)
        magsPsf.append(magPsf)
        errsPsf.append(errPsf)
        x = m.second.getCentroid().getX()
        y = m.second.getCentroid().getY()
        xCoord.append(x)
        yCoord.append(y)
        ccdNum.append(ccd)
        
        # Iterate over aperture radii
        for ir in range(10) :
            apBase = "base_CircularApertureFlux_" + str(ir)
            apFlux = apBase + "_flux"
            apFluxSig = apBase + "_fluxSigma"
            apFluxFlag = apBase + "_flag"
            diffAp.append([])
            magsAp.append([])
            errsAp.append([])
            if m.second.get(apFluxFlag) :
                # if aperture flux flag is set, set entry to large negative value
                diffAp[ir].append(-100)
                magsAp[ir].append(-100)
                errsAp[ir].append(-100)
                
            else :
                diffAp[ir].append(mag - calib.getMagnitude(m.second.get(apFlux)))
                magAp, errAp = calib.getMagnitude(m.second.get(apFlux),m.second.get(apFluxSig))
                magsAp[ir].append(magAp)
                errsAp[ir].append(errAp)

        count += 1
#    print "ccd : ",ccd," number of matches :",count
        
print "Total number of matches :", len(mags)
```

More informations on this catalog are available <A HREF=http://www.cfht.hawaii.edu/Science/CFHTLS/T0007/> here </A> and a detailed discussion about the photometric comparison between CFHTLS T0007 and SDSS DR8 is accessible <A HREF=http://www.cfht.hawaii.edu/Science/CFHTLS/T0007/T0007-docsu13.html#x43-310004.4.3> here <A>

```python
# Validation plots for aperture and psf photometry
# Apterture radii in pixels are : 3.0, 4.5, 6.0, 9.0, 12.0, 17.0, 25.0, 35.0, 50.0, 70.0

%matplotlib inline
import matplotlib
import matplotlib.pylab as plt

from lsst.meas.astrom import LoadAstrometryNetObjectsTask as Astrometry
import lsst.daf.persistence as dafPersist
import lsst.afw.geom as afwGeom
import lsst.afw.cameraGeom as cg
import lsst.afw.table as afwTable
import lsst.afw.image as afwImage
import lsst.afw.display.ds9 as ds9
import numpy
import math

# Need to specify a visit in order to retrieve the SDSS catalog corresponding to the CFHT D3 ROI
visit = 906675
filter = 'r'

# Do not consider reference magnitude > 19 
magMax = 19

# If stackCorr is True then use the color term correction implemented in the stack
stackCorr = True

#setup astrometry
solve = Astrometry.ConfigClass()
astrom = Astrometry(solve)

#setup butler - It assumes that a standard LSST directory structure is setup at the upper directory level
butler = dafPersist.Butler('../output')

# setup camera
cam = butler.get('camera')
ccd14 = cam['ccd14']
vId = {'visit':visit, 'ccd':14}
calexp = butler.get("calexp", vId, immediate=True)
detCenter = ccd14.getCenter(cg.PIXELS).getPoint()
detCenterCoord = calexp.getWcs().pixelToSky(detCenter)

# Retrieve external reference catalog (SDSS DR9)
# loadSkyCircle is replacing getReferenceSources and is returning a structure (fluxField='r_flux'; refCat)
s = astrom.loadSkyCircle(detCenterCoord, afwGeom.Angle(1., afwGeom.degrees), filter)
sdssRef = s.refCat

magCfhtU = []
magSdssU = []
magCfhtG = []
magSdssG = []
magCfhtR = []
magSdssR = []
magCfhtI = []
magSdssI = []
magCfhtZ = []
magSdssZ = []

errSdssR =[]
errCfhtR = []

#  match CFHT and SDSS to external reference catalog
match = afwTable.matchRaDec(sdssRef, cfhtRef, 1.0 * afwGeom.arcseconds)
print("%d matches found"%(len(match)))
count = 0
# Loop over matched sources
for cnts,m in enumerate(match):
    # Apply quality cuts
        if m.second.get("t_flag") != 0 or not m.first.get("photometric") :
            print "Not photometric"
            continue
            
        if m.second.get("r_mag") > 19 :
            continue
        
        magCfhtU.append(m.second.get("u_mag"))
        magCfhtG.append(m.second.get("g_mag"))
        magCfhtR.append(m.second.get("r_mag"))
        magCfhtI.append(m.second.get("i_mag"))
        magCfhtZ.append(m.second.get("z_mag"))
        
        errCfhtR.append(m.second.get("r_magSigma"))
        magErr = afwImage.abMagErrFromFluxErr(m.first.get("r_fluxSigma"),m.first.get("r_flux"))
        errSdssR.append(magErr)
        
        # Determine SDSS magnitudes from flux and apply color correction
        
        if stackCorr :
            magSdssU.append(afwImage.abMagFromFlux(m.first.get('u_flux')) - 
                            0.241*(afwImage.abMagFromFlux(m.first.get('u_flux')) - 
                                   afwImage.abMagFromFlux(m.first.get('g_flux'))))

            magSdssI.append(afwImage.abMagFromFlux(m.first.get('i_flux')) - 
                            0.085*(afwImage.abMagFromFlux(m.first.get('r_flux')) - 
                                   afwImage.abMagFromFlux(m.first.get('i_flux'))))

            magSdssR.append(afwImage.abMagFromFlux(m.first.get('r_flux')) - 
                            0.024*(afwImage.abMagFromFlux(m.first.get('g_flux')) - 
                                   afwImage.abMagFromFlux(m.first.get('r_flux'))))

            magSdssG.append(afwImage.abMagFromFlux(m.first.get('g_flux')) - 
                           0.153*(afwImage.abMagFromFlux(m.first.get('g_flux')) - 
                                   afwImage.abMagFromFlux(m.first.get('r_flux'))))

            magSdssZ.append(afwImage.abMagFromFlux(m.first.get('z_flux')) - 
                            0.074*(afwImage.abMagFromFlux(m.first.get('z_flux')) - 
                                   afwImage.abMagFromFlux(m.first.get('i_flux'))))
        else :
            magSdssU.append(afwImage.abMagFromFlux(m.first.get('u_flux')) - 
                            0.189*(afwImage.abMagFromFlux(m.first.get('u_flux')) - 
                                   afwImage.abMagFromFlux(m.first.get('g_flux'))) - 0.099)
            magSdssI.append(afwImage.abMagFromFlux(m.first.get('i_flux')) - 
                            0.102*(afwImage.abMagFromFlux(m.first.get('r_flux')) - 
                                   afwImage.abMagFromFlux(m.first.get('i_flux'))) + 0.024)
            magSdssR.append(afwImage.abMagFromFlux(m.first.get('r_flux')) - 
                            0.071*(afwImage.abMagFromFlux(m.first.get('r_flux')) - 
                                   afwImage.abMagFromFlux(m.first.get('i_flux'))) + 0.000)
            magSdssG.append(afwImage.abMagFromFlux(m.first.get('g_flux')) - 
                            0.158*(afwImage.abMagFromFlux(m.first.get('g_flux')) - 
                                   afwImage.abMagFromFlux(m.first.get('r_flux'))) +0.024)
            magSdssZ.append(afwImage.abMagFromFlux(m.first.get('z_flux')) + 
                            0.069*(afwImage.abMagFromFlux(m.first.get('i_flux')) - 
                                   afwImage.abMagFromFlux(m.first.get('z_flux'))) + 0.030)


        count += 1
        
print "Total number valid matches :", len(magCfhtR)
```

```python
font = {'family' : 'serif',
        'serif'   : 'times',
        'weight' : 'normal',
        'size'   : 15}

matplotlib.rc('font', **font)

plt.rc('text', usetex=True)

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(18, 10))

diffMag = numpy.asarray(magSdssR)-numpy.asarray(magCfhtR)
sigmaDiff = numpy.std(diffMag)
meanDiff = numpy.mean(diffMag)
ax1.hist(diffMag, bins=100, color="blue")
ax1.set_xlabel("$\Delta$ mag (SDSS - CFHT)", fontsize=20)
ax1.set_title("$\sigma$ = %.0f mmag \nmean = %.1f mmag"%(sigmaDiff*1000, meanDiff*1000), x=0.7, y=0.9, fontsize=20)
ax1.tick_params(labelsize=20)
ax1.tick_params(labelsize=20)

ax2.scatter(magSdssR, magCfhtR, s=0.5, color="blue")
ax2.set_xlabel("SDSS magnitude", fontsize=20)
ax2.set_ylabel("CFHT magnitude", fontsize=20)
ax2.tick_params(labelsize=20)

fig.tight_layout()
```
